```python
import pandas as pd
import numpy as np

# load csv as dataframe
partial = pd.read_csv('/home/jgunn/neurosketch/partial_sketch_full.csv')
partial.head()
```

```python
# match subj and time_point values to logistic_timeseries_drawing_neural_{}_{}.csv formatting
partial['wID'] = [int(each[:each.find('_')]) for each in partial['wID'].values]
partial['numSketch'] = [int(each)+1 for each in partial['numSketch'].values]

# generate t1_prob, t2_prob, c1_prob, c2_prob columns
t1_prob, t2_prob, c1_prob, c2_prob = [], [], [], []
for i in range(len(partial['index'])):
    trained = sorted((partial['target'][i], partial['competitor'][i]))
    control = sorted(set(['bed', 'bench', 'chair', 'table']) - set(trained))
    t1_prob, t2_prob = t1_prob + [partial[trained[0]][i]], t2_prob + [partial[trained[1]][i]]
    c1_prob, c2_prob = c1_prob + [partial[control[0]][i]], c2_prob + [partial[control[1]][i]]
partial['t1_prob'], partial['t2_prob'], partial['c1_prob'], partial['c2_prob'] = np.log(t1_prob), np.log(t2_prob), np.log(c1_prob), np.log(c2_prob)

# generate run_num column
partial['run_num'] = pd.cut(partial['index'].values, 4, labels=[1, 2, 3, 4])

# drop columns not found currently useful 
partial = partial.drop([i for i in partial.columns if 'Unnamed' in i or 'viewpoint' in i or 'trial' in i or 'competitor' in i] , axis=1)

# rename columns to match logistic_timeseries_drawing_neural_{}_{}.csv formatting 
partial = partial.rename(columns=
                         {'wID':'subj', 'index':'trial_num', 
                          'numSketch': 'time_point', 'target': 'label',
                          'bed': 'bed_prob', 'bench': 'bench_prob',
                          'chair': 'chair_prob', 'table': 'table_prob'})

partial.to_csv('formatted_partial_sketch.csv')
partial
# missing TR_num, which i never use
# missing ROI, which isnt desired
# missing run_num, needed
```
